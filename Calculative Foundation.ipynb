### Load Dataset First

```python

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("student_performance_100_rows_with_names.csv")

df 
```

```python
subjects = ["Math", "Physics", "Chemistry", "English", "Computer_Science"]
data = df[subjects].values
print(data)
```

### Represent each student’s scores as a vector

```python
v1 = data[0]
print(v1)
```

### Compute Norm-1 and Norm-2

```python
l1_norm = np.linalg.norm(v1, ord=1)
print("L1 Norm:", l1_norm)
l2_norm = np.linalg.norm(v1, ord=2)
print("L2 Norm:", l2_norm)
```

### Dot Product

```python
v2 = data[1]

dot_product = np.dot(v1, v2)
print("Dot Product:", dot_product)
```

### Cross Product

```python
v1_3d = v1[:3]
v2_3d = v2[:3]

cross_product = np.cross(v1_3d, v2_3d)
print("Cross Product:", cross_product)
```

### Projection of One Vector onto Another

```python
projection = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
print("Projection of v1 on v2:", projection)
```

## PART B: Matrix Operations

```python
A = data
print("Matrix Shape:", A.shape)
```

### Matrix Addition

```python
B = A.copy()
addition = A + B
print("Matrix Addition (A + B):")
print(addition[:5])
```

### Matrix Multiplication

```python
multiplication = np.dot(A, A.T)
print("Matrix Multiplication (A x A^T) Shape:", multiplication.shape)
```

### Transpose

```python
transpose = A.T
print("Transpose Shape:", transpose.shape)
```

### Determinant

```python
square_matrix = A[:5, :5]

det = np.linalg.det(square_matrix)
print("Determinant:", det)
```

### Inverse

```python
if det != 0:
    inverse = np.linalg.inv(square_matrix)
    print("Inverse Matrix:")
    print(inverse)
else:
    print("Matrix is singular, inverse not possible")
```

### PART C – Linear Transformations 

### Scaling Transformation

```python
X = df[subjects].values
A = 1.1 * np.identity(5)
X_scaled = X @ A

print(X_scaled[:5])
```

### Rotation Transformation

```python
theta = np.radians(30)

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

X_2D = X[:, :2]  # Math & Physics
X_rotated = X_2D @ R

print(X_rotated[:5])
```

### PART D – Eigenvalues + SVD

### Covariance Matrix

```python
X_standardized = StandardScaler().fit_transform(X)
cov_matrix = np.cov(X_standardized.T)

print(cov_matrix)
```

### Eigenvalues & Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
```

### Singular Value Decomposition

```python
U, S, VT = np.linalg.svd(X_standardized)

print("Singular Values:\n", S)
```

### PART E – PCA + LDA

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standardized)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

### LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

y = df["Performance_Category"]

lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(X_standardized, y)

print(X_lda[:5])
```
