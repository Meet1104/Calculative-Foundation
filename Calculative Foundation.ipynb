### Load Dataset First

```python
import numpy as np
import pandas as pd
df = pd.read_csv("student_performance_100_rows_with_names.csv")

df 
```

```python
subjects = ["Math", "Physics", "Chemistry", "English", "Computer_Science"]
data = df[subjects].values
print(data)
```

### Represent each studentâ€™s scores as a vector

```python
v1 = data[0]
print(v1)
```

### Compute Norm-1 and Norm-2

```python
l1_norm = np.linalg.norm(v1, ord=1)
print("L1 Norm:", l1_norm)
l2_norm = np.linalg.norm(v1, ord=2)
print("L2 Norm:", l2_norm)
```

### Dot Product

```python
v2 = data[1]

dot_product = np.dot(v1, v2)
print("Dot Product:", dot_product)
```

### Cross Product

```python
v1_3d = v1[:3]
v2_3d = v2[:3]

cross_product = np.cross(v1_3d, v2_3d)
print("Cross Product:", cross_product)
```

### Projection of One Vector onto Another

```python
projection = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
print("Projection of v1 on v2:", projection)
```

## PART B: Matrix Operations

```python
A = data
print("Matrix Shape:", A.shape)
```

### Matrix Addition

```python
B = A.copy()
addition = A + B
print("Matrix Addition (A + B):")
print(addition[:5])
```

### Matrix Multiplication

```python
multiplication = np.dot(A, A.T)
print("Matrix Multiplication (A x A^T) Shape:", multiplication.shape)
```

### Transpose

```python
transpose = A.T
print("Transpose Shape:", transpose.shape)
```

### Determinant

```python
square_matrix = A[:5, :5]

det = np.linalg.det(square_matrix)
print("Determinant:", det)
```

### Inverse

```python
if det != 0:
    inverse = np.linalg.inv(square_matrix)
    print("Inverse Matrix:")
    print(inverse)
else:
    print("Matrix is singular, inverse not possible")
```
